### importing necessary library

```python
import os
import numpy as np
import json
from datetime import datetime
```

### uploading our inventory 

```python
fd = open("inventory.txt",'r')

inv_txt = fd.read()

fd.close()
```

```python
inv_dict = json.loads(inv_txt)
inv_dict
```

### Customer Purchase

```python
print('Please enter 5 items and quantity key which you would like to purchase')
order={}
for i in range (0,5):
    x=input()
    tmp=x.split(' ')
    order[tmp[0]] = int(tmp[1])
          
```

```python
print(order)
```

```python
type(order)
```

### updating inventory

```python
for i in order:    
    if(i in inv_dict.keys()):
        print('you ordered {} {}'.format(order[i],inv_dict[i]['name']))
        print('old quantity is',inv_dict[i]['quantity'])
        inv_dict[i]['quantity'] -= order[i]
        print('new quantity is',inv_dict[i]['quantity'])
    else:
        print("product {} is not available".format(i))
        
```

```python
inv_dict
```

### total bill

```python
cost=0
for i in order:    
    if(i in inv_dict.keys()):
        print('you ordered {} {}'.format(order[i],inv_dict[i]['name']))
        print('Cost {} '.format(order[i]*inv_dict[i]['price']))
        cost+=order[i]*inv_dict[i]['price']
print('Total bill : ',cost)

bill={}
c=1
for i in order:
    if(i in inv_dict.keys()):
        yo={}
        yo['item'] = inv_dict[i]['name']
        yo['quantity'] = order[i]
        yo['price'] = inv_dict[i]['price']
        yo['SP'] = inv_dict[i]['price']*(1.0 - inv_dict[i]['discount']/100.0)
        yo['total'] = yo['SP']*order[i]
        bill[c] = yo
        c+=1
print(bill)
```

### saving json file

```python
js = json.dumps(bill)

fd = open("sales.txt",'w')

fd.write(js)

fd.close()
```
